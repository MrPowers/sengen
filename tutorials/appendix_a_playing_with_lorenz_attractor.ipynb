# Appendix A: Playing With Lorenz Attractor

## A. Plotting X(i), X(i+1), and X(i+2)

```python
%matplotlib inline
from data_generator import SensorDataGenerator as sdg
dg = sdg()
dg.generation_input.add_option(sensor_names="x y z", attractor="lorenz", rho=28.0, sigma=10.0, beta=8.0/3.0,
                             initial_state=[0.2,0.2,0.2], anomaly=0.0, time_interval=0.2)
dg.generate(sample_size=2000)
dg.plot_data()
```

```python
import numpy as np
new_table = {}
new_table["x"] = np.array(dg.data["x"][:-2])
new_table["x+dt"]  = np.array(dg.data["x"][1:-1])
new_table["x+2dt"]  = np.array(dg.data["x"][2:])

import pandas as pd
df = pd.DataFrame(new_table)
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt 
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.set_title("")
ax.set_xlabel("x")
ax.set_ylabel("x+dt")
ax.set_zlabel("x+2dt")
ax.plot(df["x"], df["x+dt"], df["x+2dt"], color="blue", lw=0.1)
plt.show()
```

## Plotting X(i) and X(i+1)

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt 
plt.plot(df["x"], df["x+dt"], 'o-', alpha=0.2)
plt.show()
```

## C. Lorenz Attractor w/ Rho=99.96, Sigma=10.0, Beta=2.6666...

```python
from data_generator import SensorDataGenerator as sdg
dg = sdg()
dg.generation_input.add_option(sensor_names="x y z", attractor="lorenz", rho=99.96, sigma=10.0, beta=8.0/3.0,
                             initial_state=[0.2,0.2,0.2], anomaly=0.0, time_interval=0.2)
dg.generate(sample_size=2000)
df=dg.data
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt 
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.set_title("")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
ax.plot(df["x"], df["y"], df["z"], color="blue", lw=0.1)
plt.show()
plt.plot(df["x"], df["y"], 'o', alpha=0.2)
plt.show()
```
